##  Bivariate Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
titanic.head()
```

```python
flights.head()
```

```python
iris.head()
```

### 1. Scatter Plots (Num-Num)

```python
sns.scatterplot(tips['total_bill'], tips['tip'])
```

```python
tips.head()
```

```python
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['sex'])
```

```python
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['sex'], style = tips['smoker'])
```

```python
sns.scatterplot(tips['total_bill'], tips['tip'], hue=tips['sex'], size = tips['size'])
```

### 2. Barplots (Num-Cat)

```python
sns.barplot(titanic['Pclass'], titanic['Age'])
```

```python
sns.barplot(titanic['Pclass'], titanic['Age'], hue = titanic['Sex'])
```

### 3. Box Plot (Num-Cat)

```python
sns.boxplot(titanic['Age'], titanic['Sex'])
```

```python
sns.boxplot(titanic['Age'], titanic['Sex'], hue = titanic['Survived'])
```

### 4. DistPlot(Num-Cat)

```python
sns.distplot(titanic['Age'])
```

### 5.HeatMap(Cat-Cat)

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

### 6. ClusterMap (Cat-Cat)

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

### 7. Pairplot

```python
sns.pairplot(iris, hue = 'species')
```

### 8. Lineplot (Num-Num)


```python
new = flights.groupby('year').sum().reset_index()
new
```

```python
sns.lineplot(new['year'], new['passengers'])
```
